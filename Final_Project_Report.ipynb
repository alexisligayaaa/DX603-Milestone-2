### 4.2 Data Cleaning and Preprocessing

Describe how you prepared the data.

Include:
- issues in the raw data
- handling of missing values, outliers, inconsistencies
- key decisions and why
- what worked and what did not work

**4.2 Answer:**  
The raw dataset had several issues including missing values, redundant or irrelevant features, and extreme outliers. Many columns, such as basement and building type features, had over 90% missing data, making them unreliable for modeling. We also removed features that lacked meaningful predictive value, such as identifier variables, like parcelid, and encoded location IDs, such as regionidzip and regionidcity, because they do not directly represent property characteristics.

To address these issues, we removed features with little variation or excessive missing data. Rows with missing target values were dropped since they could not be used for training. We also removed rows with more than 40% missing data to improve quality. Outliers in the target variable were handled using the IQR method to reduce the impact of extreme property values.

For the remaining missing values, numerical features were filled using the median to reduce the impact of outliers. Categorical features were also filled with the most common value and then converted into numerical form using ordinal encoding so they could be used in the models.

Overall, these preprocessing steps improved the consistency and usability of the dataset. Cleaning the data clearly improved model stability. More extensive feature engineering later did not significantly improve performance, suggesting that data quality mattered more than adding new features.
