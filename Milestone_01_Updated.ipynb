### Part 3.C: Drop Problematic **samples** 

There could be several reasons why you might want to drop a sample:

```python
df_3c = df_3b.copy()
rows_before = len(df_3c)

df_3c = df_3c.dropna(subset=['taxvaluedollarcnt'])
row_missing_fraction = df_3c.isnull().mean(axis=1)
df_3c = df_3c.loc[row_missing_fraction <= 0.40].copy()

q1, q3 = df_3c['taxvaluedollarcnt'].quantile([0.25, 0.75])
iqr = q3 - q1
lower_bound = max(0, q1 - 3 * iqr)
upper_bound = q3 + 3 * iqr

df_3c = df_3c[(df_3c['taxvaluedollarcnt'] >= lower_bound) & (df_3c['taxvaluedollarcnt'] <= upper_bound)].copy()

rows_after = len(df_3c)
print(f'Rows before filtering: {rows_before:,}')
print(f'Rows after filtering:  {rows_after:,}')
print(f'Rows dropped:          {rows_before - rows_after:,}')
```

We cleaned the rows in three simple steps. First, we removed rows where `taxvaluedollarcnt` was missing, because we need the target value to train a model. Second, we removed rows that were missing more than 40% of their values, since those rows did not have enough useful information. Third, we removed very extreme target values using the IQR method so unusual outliers would not overly affect the results. This keeps most normal data while reducing noise.

```python
# Log Transformation 
from sklearn.preprocessing import FunctionTransformer

def log_transformation(df: pd.DataFrame, feature: str) -> pd.DataFrame:

    if feature not in df.columns:
        raise ValueError(f"Column '{feature}' not found in DataFrame.")

    if (df[feature] < 0).any():
        raise ValueError(
            f"Column '{feature}' contains negative values")
        

    log_transformer = FunctionTransformer(np.log1p, validate=False)

    df = df.copy()
    df[f"{feature}_log"] = log_transformer.transform(df[[feature]])

    return df
```

```python
#Ration Trasnformation Function
def ratio_features(
    df: pd.DataFrame,
    numerator: str,
    denominator: str,
    new_feature_name: str | None = None
) -> pd.DataFrame:

    if numerator not in df.columns:
        raise ValueError(f"Column '{numerator}' not found in DataFrame.")
    if denominator not in df.columns:
        raise ValueError(f"Column '{denominator}' not found in DataFrame.")

    if new_feature_name is None:
        new_feature_name = f"{numerator}_to_{denominator}"

    def safe_ratio(X):
        num = X.iloc[:, 0]
        den = X.iloc[:, 1]

        ratio = np.where(den == 0, np.nan, num / den)
        return ratio.reshape(-1, 1)

    ratio_transformer = FunctionTransformer(safe_ratio, validate=False)

    df = df.copy()
    df[new_feature_name] = ratio_transformer.transform(df[[numerator, denominator]])

    return df
```

```python
#Run the Ratio feature transformation
features = X.columns
df_new = X.copy()
for numerator in features:
    for denominator in features:
        if numerator != denominator:
            new_feature_name = f"{numerator}_to_{denominator}"
            df_new = ratio_features(
                df_new,
                numerator=numerator,
                denominator=denominator,
                new_feature_name=new_feature_name
            )
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def scale_features(
    df: pd.DataFrame,
    features: list[str],
    method: str = "standard"
) -> pd.DataFrame:
    
    df = df.copy()

    for feature in features:
        if feature not in df.columns:
            raise ValueError(f"Column '{feature}' not found in DataFrame.")
        if not pd.api.types.is_numeric_dtype(df[feature]):
            raise ValueError(f"Column '{feature}' must be numeric.")

    if method == "standard":
        scaler = StandardScaler()
        suffix = "_scaled"
    elif method == "minmax":
        scaler = MinMaxScaler()
        suffix = "_normalized"
    else:
        raise ValueError("method must be either 'standard' or 'minmax'")

    scaled_values = scaler.fit_transform(df[features])

    scaled_column_names = [f"{feature}{suffix}" for feature in features]

    df[scaled_column_names] = scaled_values

    return df
```

## Appendix 2: Summary of Basic Imputation Methods
Imputation depends on the data type and context. Below are common techniques for handling missing values.

---

### **1. Simple Imputation (Basic Methods)**
| Method | Best For | Code Example | Pros | Cons |
|--------|---------|--------------|------|------|
| **Drop Missing Values** | Few missing values (<5% of data) | `df.dropna()` | Quick and easy | Can remove valuable data |
| **Mean Imputation** | Normally distributed numerical data | `df.fillna(df.mean())` | Preserves mean; simple | Distorts variance, weak for skewed data |
| **Median Imputation** | Skewed numerical data | `df.fillna(df.median())` | Robust to outliers | May not capture patterns |
| **Mode Imputation** | Categorical features | `df.fillna(df.mode().iloc[0])` | Keeps most common category | Can introduce bias |
| **Constant Value (e.g., 0)** | Special cases (e.g., unknown numerical data) | `df.fillna(0)` | Simple and interpretable | Can mislead model |
| **"Unknown" Category Imputation** (**New Addition**) | Categorical features with missing values | `df.fillna('Unknown')` | Keeps all rows, prevents data loss | May introduce artificial category |

---

### **2. Statistical & Advanced Imputation**
| Method | Best For | Code Example | Pros | Cons |
|--------|---------|--------------|------|------|
| **Interpolation** | Time series, ordered data | `df.interpolate(method='linear')` | Preserves trends | May not work for non-continuous data |
| **K-Nearest Neighbors (KNN)** | Small datasets, patterns in features | `KNNImputer(n_neighbors=5).fit_transform(df)` | Uses similar observations | Computationally expensive |
| **Multivariate Imputation (MICE)** | Complex relationships between variables | `IterativeImputer().fit_transform(df)` | Captures relationships | Slower than mean/median |
| **Regression Imputation** | When missing values depend on other variables | Train regression model to predict missing values | More accurate than mean/median | Risk of overfitting |

---

### **When to Use "Unknown" Category for Categorical Imputation**
✅ **Good for:**
- Categorical features where missing values may indicate meaningful differences.  
- Customer data (e.g., missing survey responses → "No Response").  
- Product categories where missing could be a separate group.  

❌ **Avoid if:**
- The missing category does **not** have a meaningful interpretation.
- The model might learn spurious patterns from an artificial category.

---

### **Which Method to Choose?**
| Scenario | Best Method |
|----------|------------|
| **Few missing values (<5%)** | Drop NaNs (`df.dropna()`) |
| **Numerical & normal distribution** | Mean (`df.fillna(df.mean())`) |
| **Numerical & skewed distribution** | Median (`df.fillna(df.median())`) |
| **Categorical features** | Mode (`df.fillna(df.mode().iloc[0])`) |
| **Categorical with possible meaning in missingness** | "Unknown" Category (`df.fillna('Unknown')`) |
| **Small dataset with patterns** | KNN Imputer (`KNNImputer()`) |
| **Complex relationships between features** | MICE / Iterative Imputer |
